```python
import pandas as pd
import os
import glob
#glob documentation here and here:
##https://docs.python.org/3/library/glob.html
###https://stackoverflow.com/questions/37080132/use-pandas-and-glob-to-merge-numerous-csv-data-text-files-from-a-directory
```

```python
#point to the folder housing the data
path = "CitiBikeData"

#glob all the csvs
all_csvs = glob.glob(os.path.join(path, "*csv"))

#make dataframes for all the csv files
df_all = (pd.read_csv(f) for f in all_csvs)

#concatenate the dataframes
concatenated_df = pd.concat(df_all, ignore_index=True)
concatenated_df.head()
```

```python
#output the concatenated data to a csv
concatenated_df.to_csv("CitiBikeData/citibike_567_2021.csv")
```
